```python
import pandas as pd
import numpy as np
import re 
```

```python
# Î§ÏÎ·ÏƒÎ¹Î¼Î¿Ï€Î¿Î¹Î¿ÏÎ¼Îµ Ï„Î·Î½ Ï€Î±ÏÎ¬Î¼ÎµÏ„ÏÎ¿ lineterminator Î³Î¹Î±Ï„Î¯ Ï‡Ï‰ÏÎ¯Ï‚ Î±Ï…Ï„Î®Î½ Ï„Î± Î´ÎµÎ´Î¿Î¼Î­Î½Î± Î´ÎµÎ½ "Î´Î¹Î±Î²Î¬Î¶Î¿Î½Ï„Î±Î¹" Î¿ÏÎ¸Î¬. 
# ÎŸ Î»ÏŒÎ³Î¿Ï‚ ÎµÎ¯Î½Î±Î¹ Î¯ÏƒÏ‰Ï‚ ÏŒÏ„Î¹ Ï„Î± CSV Î±ÏÏ‡ÎµÎ¯Î± Î´Î·Î¼Î¹Î¿Ï…ÏÎ³Î®Î¸Î·ÎºÎ±Î½ ÏƒÎµ Ï…Ï€Î¿Î»Î¿Î³Î¹ÏƒÏ„Î® Î¼Îµ Î»ÎµÎ¹Ï„Î¿Ï…ÏÎ³Î¹ÎºÏŒ ÏƒÏÏƒÏ„Î·Î¼Î± Linux.
tweets = pd.read_csv('Data/tweets/tweets_2021_with_users.csv', lineterminator='\n')
records = pd.read_csv('Data/NVD/NVD_records_2022_downoload_at_February_2022_01-02-2022.csv', lineterminator='\n')
```

Î£Ï„Î·Î½ ÏƒÏ…Î½Î­Ï‡ÎµÎ¹Î± Ï€ÎµÏÎ¹Î³ÏÎ¬Ï†ÎµÏ„Î±Î¹ Î¿ Î±Î»Î³ÏŒÏÎ¹Î¸Î¼Î¿Ï‚ Ï€Î¿Ï… Ï…Î»Î¿Ï€Î¿Î¹ÎµÎ¯ Ï„Î¿ matching. 

Î“Î¹Î± ÎºÎ¬Î¸Îµ Î­Î½Î± Î±Ï€ÏŒ Ï„Î± tweets ÎµÎºÏ„ÎµÎ»Î¿ÏÎ½Ï„Î±Î¹ Ï„Î± Î±ÎºÏŒÎ»Î¿Ï…Î¸Î± Î²Î®Î¼Î±Ï„Î±:
1. Î§ÏÎ·ÏƒÎ¹Î¼Î¿Ï€Î¿Î¹ÏÎ½Ï„Î±Ï‚ regular expression, ÎµÎ½Ï„Î¿Ï€Î¯Î¶Î¿Ï…Î¼Îµ Î±ÏÏ‡Î¹ÎºÎ¬ ÏŒÎ»Î± Ï„Î± CVE IDs Ï€Î¿Ï… Ï…Ï€Î¬ÏÏ‡Î¿Ï…Î½ ÏƒÏ„Î¿ ÎºÎµÎ¯Î¼ÎµÎ½Î¿ Ï„Î¿Ï… tweet.
2. Î£Ï„Î·Î½ ÏƒÏ…Î½Î­Ï‡ÎµÎ¹Î±, Î³Î¹Î± ÎºÎ¬Î¸Îµ Î­Î½Î± Î±Ï€ÏŒ Î±Ï…Ï„Î¬ Ï„Î± IDs Ï€Î¿Ï… Î²ÏÎ®ÎºÎ±Î¼Îµ ÏƒÏ„Î¿ Ï€ÏÎ¿Î·Î³Î¿ÏÎ¼ÎµÎ½Î¿ Î²Î®Î¼Î±, ÎºÏÎ±Ï„Î¬Î¼Îµ ÏƒÎµ Î¼Î¹Î± Î»Î¯ÏƒÏ„Î± Ï„Î± Î­Î³ÎºÏ…ÏÎ± IDs, Î´Î·Î»Î±Î´Î® Ï„Î± IDs Ï€Î¿Ï… Ï…Ï€Î¬ÏÏ‡Î¿Ï…Î½ Ï‰Ï‚ ÎµÎ³Î³ÏÎ±Ï†Î­Ï‚ ÏƒÏ„Î¿ NVD DataFrame Î¼Î±Ï‚ Ï„Î¿ Î¿Ï€Î¿Î¯Î¿ Ï€ÎµÏÎ¹Î­Ï‡ÎµÎ¹ Ï„Î¹Ï‚ ÎµÎ³Î³ÏÎ±Ï†Î­Ï‚ Î¼Îµ Ï„Î± ÎµÏ€Î¯ÏƒÎ·Î¼Î± Î´ÎµÎ´Î¿Î¼Î­Î½Î± ÎµÏ…Ï€Î±Î¸ÎµÎ¹ÏÎ½ Î±Ï€ÏŒ Ï„Î·Î½ NVD.
3. Î ÏÎ¿ÏƒÎ¸Î­Ï„Î¿Ï…Î¼Îµ ÏƒÏ„Î·Î½ Î»Î¯ÏƒÏ„Î± data Ï„Î± Î­Î³ÎºÏ…ÏÎ± CVE IDs Ï€Î¿Ï… Î²ÏÎ®ÎºÎ±Î¼Îµ ÏƒÎµ Î±Ï…Ï„ÏŒ Ï„Î¿ tweet. Î— Î»Î¯ÏƒÏ„Î± Î±Ï…Ï„Î® Ï€ÎµÏÎ¹Î­Ï‡ÎµÎ¹ Î³Î¹Î± ÎºÎ¬Î¸Îµ tweet Î¼Î¹Î± Î»Î¯ÏƒÏ„Î± Î¼Îµ Ï„Î± Î­Î³ÎºÏ…ÏÎ± CVE IDs Ï€Î¿Ï… Î±Î½Î±Ï†Î­ÏÎ¿Î½Ï„Î±Î¹ ÏƒÎµ Î±Ï…Ï„ÏŒ Î® NaN ÎµÎ¬Î½ ÏƒÏ„Î¿ ÏƒÏ…Î³ÎºÎµÎºÏÎ¹Î¼Î­Î½Î¿ tweet Î´ÎµÎ½ Î±Î½Î±Ï†Î­ÏÎµÏ„Î±Î¹ ÎºÎ±Î½Î­Î½Î± Î­Î³ÎºÏ…ÏÎ¿ CVE ID.

Î— Î»Î¯ÏƒÏ„Î± data Ï€Î¿Ï… ÎµÎ¯Î½Î±Î¹ Î· Î­Î¾Î¿Î´Î¿Ï‚ Ï„Î·Ï‚ ÎµÎºÏ„Î­Î»ÎµÏƒÎ·Ï‚ Ï„Î¿Ï… Ï€Î±ÏÎ±Ï€Î¬Î½Ï‰ Î±Î»Î³ÏŒÏÎ¹Î¸Î¼Î¿Ï…, ÎµÎ¯Î½Î±Î¹ Ï€Î±ÏÎ¬Î»Î»Î·Î»Î· Î»Î¯ÏƒÏ„Î± Î¼Îµ Ï„Î·Î½ texts, Î· Î¿Ï€Î¿Î¯Î± Ï€ÎµÏÎ¹Î­Ï‡ÎµÎ¹ Ï„Î¿ ÎºÎµÎ¯Î¼ÎµÎ½Î¿ Ï„Î¿Ï… ÎºÎ¬Î¸Îµ tweet.

Î˜Ï…Î¼Î¯Î¶Î¿Ï…Î¼Îµ ÏŒÏ„Î¹ Î´ÎµÎ½ Ï…Ï€Î¬ÏÏ‡Î¿Ï…Î½ CVE IDs Î³Î¹Î± ÎµÏ…Ï€Î¬Î¸ÎµÎ¹ÎµÏ‚ Ï€Î¿Ï… ÎµÎ¯Î½Î±Î¹ RESERVED, REJECT, DISPUTED, UNVERIFIABLE ÎºÎ±Î¹ PRERELEASE.
ÎŸÏ€ÏŒÏ„Îµ ÎµÎ¬Î½ Î¿ Î±Î»Î³ÏŒÏÎ¹Î¸Î¼Î¿Ï‚ ÏƒÏ…Î½Î±Î½Ï„Î¯ÏƒÎµÎ¹ ÏƒÏ„Î¿ ÎºÎµÎ¯Î¼ÎµÎ½Î¿ ÎµÎ½ÏŒÏ‚ tweet Ï„Î­Ï„Î¿Î¹Î¿Ï… ÎµÎ¯Î´Î¿Ï…Ï‚ CVE IDs Î´ÎµÎ½ Ï„Î± Î¸ÎµÏ‰ÏÎµÎ¯ Î­Î³ÎºÏ…ÏÎ±.

```python
# ÎšÏÎ±Ï„Î¬Î¼Îµ ÏƒÏ„Î·Î½ Î»Î¯ÏƒÏ„Î± texts Ï„Î·Î½ ÏƒÏ„Î®Î»Î· Î¼Îµ Ï„Î± ÎºÎµÎ¯Î¼ÎµÎ½Î± Ï„Ï‰Î½ tweets ÎºÎ±Î¹ ÏƒÏ„Î·Î½ Î»Î¯ÏƒÏ„Î± nvd_ids Ï„Î± CVE IDs Î±Ï€ÏŒ Ï„Î·Î½ NVD.
texts = list(tweets['text'])  
nvd_ids = list(records['cve_id'])
data = [] 
```

```python
# Î§ÏÎ·ÏƒÎ¹Î¼Î¿Ï€Î¿Î¹Î¿ÏÎ¼Îµ Î­Î½Î±Î½ Î¼ÎµÏ„ÏÎ·Ï„Î® Î³Î¹Î± Î½Î± ÎµÎ¼Ï†Î±Î½Î¯Î¶Î¿Ï…Î¼Îµ Î±Î½Î¬ 10.000 Ï„Î± tweets Ï€Î¿Ï… Î­Ï‡Î¿Ï…Î¼Îµ Ï€ÏÎ¿ÏƒÏ€ÎµÎ»Î¬ÏƒÎµÎ¹.
i = 0 

for text in texts:
    if (i % 10000 == 0):
        print("ÎˆÏ‰Ï‚ Ï„ÏÏÎ± Î­Ï‡Î¿Ï…Î½ Ï€ÏÎ¿ÏƒÏ€ÎµÎ»Î±ÏƒÏ„ÎµÎ¯", i, "tweets.")
    i = i +1   
    
    try:
        
        # Î— valid_matched_ids ÎµÎ¯Î½Î±Î¹ Î· Ï…Ï€Î¿Î»Î¯ÏƒÏ„Î± ÏƒÏ„Î·Î½ Î¿Ï€Î¿Î¯Î± Î±Ï€Î¿Î¸Î·ÎºÎµÏÎ¿Î½Ï„Î±Î¹ Ï„Î± CVE IDs Ï€Î¿Ï… Î±Î½Î±Ï†Î­ÏÎ¿Î½Ï„Î±Î¹ ÏƒÏ„Î¿ ÎºÎµÎ¯Î¼ÎµÎ½Î¿
        # Ï„Î¿Ï… ÏƒÏ…Î³ÎºÎµÎºÏÎ¹Î¼Î­Î½Î¿Ï… tweet ÎºÎ±Î¹ Î­Ï‡Î¿Ï…Î½ ÎµÎ»ÎµÎ³Ï‡Î¸ÎµÎ¯ ÏŒÏ„Î¹ Ï€ÎµÏÎ¹Î»Î±Î¼Î²Î¬Î½Î¿Î½Ï„Î±Î¹ ÏƒÏ„Î·Î½ NVD.
        valid_matched_ids = []      
        
        # ÎœÎµ Ï„Î·Î½ findall Î²ÏÎ¯ÏƒÎºÎ¿Ï…Î¼Îµ Ï„Î± CVE IDs Ï€Î¿Ï… Î±Î½Î±Ï†Î­ÏÎ¿Î½Ï„Î±Î¹ ÏƒÏ„Î¿ tweet Ï‡ÏÎ·ÏƒÎ¹Î¼Î¿Ï€Î¿Î¹ÏÎ½Ï„Î±Ï‚ regular expression.
        # Î•Î¯Î½Î±Î¹ ÏŒÎ¼Ï‰Ï‚ Î±Ï€Î±ÏÎ±Î¯Ï„Î·Ï„Î¿Ï‚ Î¿ Î­Î»ÎµÎ³Ï‡Î¿Ï‚ Ï€Î¿Ï… 
        # Î³Î¯Î½ÎµÏ„Î±Î¹ ÏƒÏ„Î·Î½ ÏƒÏ…Î½Î­Ï‡ÎµÎ¹Î± Î­Ï„ÏƒÎ¹ ÏÏƒÏ„Îµ Î½Î± Î¼Î·Î½ ÎºÏÎ±Ï„Î®ÏƒÎ¿Ï…Î¼Îµ Î¼Î· Î­Î³ÎºÏ…ÏÎ± CVE IDs (Î´Î·Î». IDs Ï€Î¿Ï… Î´ÎµÎ½ Ï…Ï€Î¬ÏÏ‡Î¿Ï…Î½ ÏƒÏ„Î¿
        # dataset Ï€Î¿Ï… Ï†Ï„Î¹Î¬Î¾Î±Î¼Îµ Î±Ï€ÏŒ Ï„Î± ÎµÏ€Î¯ÏƒÎ·Î¼Î± Î´ÎµÎ´Î¿Î¼Î­Î½Î± Ï„Î·Ï‚ NVD). Î— findall ÎµÏ€Î¹ÏƒÏ„ÏÎ­Ï†ÎµÎ¹ Î¼Î¹Î± Î»Î¯ÏƒÏ„Î± Î¼Îµ Ï„Î± 
        # Î±Î»Ï†Î±ÏÎ¹Î¸Î¼Î·Ï„Î¹ÎºÎ¬ Ï€Î¿Ï… Ï„Î±Î¹ÏÎ¹Î¬Î¶Î¿Ï…Î½ ÏƒÏ„Î¿ pattern.
        #
        # Î— findall ÎµÏ€Î¹ÏƒÏ„ÏÎ­Ï†ÎµÎ¹ Î¼Î¹Î± Î»Î¯ÏƒÏ„Î± Î¼Îµ Ï„Î± strings Ï€Î¿Ï… Ï„Î±Î¹ÏÎ¹Î¬Î¶Î¿Ï…Î½ ÏƒÏ„Î¿ pattern Ï€Î¿Ï… Î­Ï‡Î¿Ï…Î¼Îµ Î´Î·Î»ÏÏƒÎµÎ¹, ÎºÎ±Î¹ Î±Ï…Ï„ÏŒ ÏƒÎ·Î¼Î±Î¯Î½ÎµÎ¹
        # ÏŒÏ„Î¹ ÎµÏ€Î¹Ï„ÏÎ­Ï€ÎµÎ¹ Ï„Î± Î´Î¹Ï€Î»ÏŒÏ„Ï…Ï€Î±. ÎˆÏ„ÏƒÎ¹ ÎµÎ¬Î½ ÏƒÎµ Î­Î½Î± tweet Î­Ï‡Î¿Ï…Î¼Îµ Î­Î½Î± ID Ï€Î¿Ï… Î±Î½Î±Ï†Î­ÏÎµÏ„Î±Î¹ Î´ÏÎ¿ Ï†Î¿ÏÎ­Ï‚, Î¸Î± Î±Ï€Î¿Î¸Î·ÎºÎµÏ…Ï„ÎµÎ¯
        # ÏƒÏ„Î· Î»Î¯ÏƒÏ„Î± Î´ÏÎ¿ Ï†Î¿ÏÎ­Ï‚. Î•Î¼ÎµÎ¯Ï‚ ÏŒÎ¼Ï‰Ï‚ Î¸Î­Î»Î¿Ï…Î¼Îµ Î±Ï€ÏŒ ÎºÎ¬Î¸Îµ tweet Î½Î± ÎºÏÎ±Ï„Î¬Î¼Îµ Ï„Î± Î¼Î¿Î½Î±Î´Î¹ÎºÎ¬ IDs Ï€Î¿Ï… Î±Î½Î±Ï†Î­ÏÎ¿Î½Ï„Î±Î¹ ÏƒÎµ Î±Ï…Ï„ÏŒ.
        # ÎŸÏ€ÏŒÏ„Îµ Î¼ÎµÏ„Î±Ï„ÏÎ­Ï€Î¿Ï…Î¼Îµ Ï„Î·Î½ Î»Î¯ÏƒÏ„Î± ÏƒÎµ set ÎºÎ¹ Î­Ï„ÏƒÎ¹ Î´Î¹Î±Î³ÏÎ¬Ï†Î¿Ï…Î¼Îµ Ï„Î± Î´Î¹Ï€Î»ÏŒÏ„Ï…Ï€Î±.
        matched_ids = set(re.findall("CVE-\d{1,6}-\d{1,6}", text))  
        
        for matched in matched_ids:
            if matched in nvd_ids:
                valid_matched_ids.append(matched)
                
        # Î•Î¬Î½ Î´ÎµÎ½ Î²ÏÎ®ÎºÎ±Î¼Îµ ÎºÎ±Î½Î­Î½Î± Î­Î³ÎºÏ…ÏÎ¿ CVE ID ÏƒÏ„Î¿ ÎºÎµÎ¯Î¼ÎµÎ½Î¿ Ï„Î¿Ï… text Ï„ÏŒÏ„Îµ ÎºÎ±Ï„Î±Ï‡ÏÏÎ·ÏƒÎµ np.NaN
        # ÏƒÏ„Î·Î½ Î¸Î­ÏƒÎ· Ï„Î·Ï‚ Î»Î¯ÏƒÏ„Î±Ï‚ data Ï€Î¿Ï… Î±Î½Ï„Î¹ÏƒÏ„Î¿Î¹Ï‡ÎµÎ¯ ÏƒÏ„Î¿ tweet Î±Ï…Ï„ÏŒ.        
        if (len(valid_matched_ids) == 0):
            data.append(np.NaN)     
   
        else:
            # Î‘Ï€Î¿Î¸Î®ÎºÎµÏ…ÏƒÎ· Ï„Ï‰Î½ Î­Î³ÎºÏ…ÏÏ‰Î½ IDs Ï€Î¿Ï… ÎµÎ½Ï„Î¿Ï€Î¯ÏƒÎ±Î¼Îµ ÏƒÏ„Î¿ ÎºÎµÎ¯Î¼ÎµÎ½Î¿.
            data.append(valid_matched_ids)        
        
    except Exception as e:
        print("ÎˆÎ½Î± ÏƒÏ†Î¬Î»Î¼Î± ÎµÎ¼Ï†Î±Î½Î¯ÏƒÏ„Î·ÎºÎµ:")
        print(e)
        break
```

```python
if (len(tweets) == len(data)):
    print("Î¤Î¿ matching Î¿Î»Î¿ÎºÎ»Î·ÏÏÎ¸Î·ÎºÎµ ÎµÏ€Î¹Ï„Ï…Ï‡ÏÏ‚! Î— Î»Î¯ÏƒÏ„Î± data Ï€ÏÎ¿ÏƒÏ„Î¯Î¸ÎµÏ„Î±Î¹ ÏƒÎ±Î½ Î½Î­Î± ÏƒÏ„Î®Î»Î· ÏƒÏ„Î¿ DataFrame tweets.")
    tweets.insert(2, 'matched_CVE_IDs', data)
else:
    print("ÎšÎ¬Ï„Î¹ Ï€Î®Î³Îµ ÏƒÏ„ÏÎ±Î²Î¬ ÎºÎ±Î¹ Ï„Î¿ Î¼Î­Î³ÎµÎ¸Î¿Ï‚ Ï„Î·Ï‚ Î»Î¯ÏƒÏ„Î±Ï‚ data Î´ÎµÎ½ ÎµÎ¯Î½Î±Î¹ Î¯ÏƒÎ¿ Î¼Îµ\n Ï„Î¿ Î¼Î­Î³ÎµÎ¸Î¿Ï‚ Ï„Î¿Ï… dataframe tweets Î¿Ï€ÏŒÏ„Îµ Î´ÎµÎ½ Î¼Ï€Î¿ÏÎ¿ÏÎ¼Îµ Î½Î± Ï„Î± ÎµÎ½ÏÏƒÎ¿Ï…Î¼Îµ...ÎÎ±Î½Î±Î´Î­Ï‚ Ï„Î¿!")
```
